```r
# Charger le package ggplot2
library(ggplot2)

# Utiliser le dataset intégré 'mpg'
data(mpg)

# Créer le graphique
p <- ggplot(mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point(size = 3) +  # Ajouter des points
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE) +  # Ajouter une courbe de lissage linéaire avec une formule explicite
  labs(title = "Relation entre la cylindrée et l'efficacité de carburant",
       x = "Cylindrée (litres)",
       y = "Efficacité de carburant (miles par gallon)",
       color = "Classe de véhicule") +
  theme_minimal()  # Utiliser un thème minimal pour une meilleure esthétique

# Afficher le graphique
print(p)

```
