```r
# Charger le package ggplot2
library(ggplot2)

# Utiliser le dataset intgr 'mpg'
data(mpg)

# Crer le graphique
p <- ggplot(mpg, aes(x = displ, y = hwy, color = class)) +
  geom_point(size = 3) +  # Ajouter des points
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE) +  # Ajouter une courbe de lissage linaire avec une formule explicite
  labs(title = "Relation entre la cylindre et l'efficacit de carburant",
       x = "Cylindre (litres)",
       y = "Efficacit de carburant (miles par gallon)",
       color = "Classe de vhicule") +
  theme_minimal()  # Utiliser un thme minimal pour une meilleure esthtique

# Afficher le graphique
print(p)

```
